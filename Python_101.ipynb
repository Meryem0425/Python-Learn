Short Circuit

 Python'da kısa devre değerlendirme mantığı, mantıksal ifadelerin sadece gerekli olduğu kadarını değerlendirerek, gereksiz işlemlerden kaçınmayı sağlar. Örneğin, "and" ifadesinde ilk ifade yanlışsa, ikinci ifadeye bakılmaz çünkü sonuç zaten yanlış olacaktır. "or" ifadesinde ise, ilk ifade doğruysa, ikinci ifadeye bakılmaz çünkü sonuç zaten doğru olacaktır. Bu şekilde, kısa devre değerlendirme, programın daha verimli ve hızlı çalışmasını sağlar.

 (5<3) and print("hey")
 output:False // Hey yazmadı çünkü 5 3 den küçük değil false döndürdü.
